```python
import pandas as pd
import numpy as np
df=pd.read_csv("/content/drive/MyDrive/students_placement_data.csv")
df.head()
```

```python
df["SSC Percentage"]=df["SSC Percentage"].replace(to_replace=np.nan,value=np.mean(df["SSC Percentage"]))
df.head()
```

```python
print(df['Gender'].value_counts())
print(df['Section'].value_counts())
print(df['placement status'].value_counts())
print(df['registered_for_ Placement_Training'].value_counts())
```

```python
import plotly.express as px
cut_counts=df['Gender'].value_counts()
fig=px.bar(x=cut_counts.index,y=cut_counts.values)
fig.show()
```

```python
pd.crosstab(index=df['Gender'],columns=df['placement status'])
```
